```python
import pandas as pd
```

```python
data=pd.read_csv("insurance_pre.csv")
```

```python
data
```

```python
data=pd.get_dummies(data,drop_first=True)
```

```python
data
```

```python
data.columns
```

```python
independent=data[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent=data[[ 'charges']]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
param_grid={"criterion":["squared_error","absolute_error", "friedman_mse"],
           "max_features":["sqrt","log2"],"n_estimators":[10,100] }

grid=GridSearchCV(RandomForestRegressor(),param_grid,refit=True, verbose=3,n_jobs=-1)
```

```python
grid.fit(independent,dependent)
```

```python
print("The best tuning Parameter {}:".format(grid.best_params_))
```

```python
results=grid.cv_results_
```

```python
results
```

```python
table=pd.DataFrame.from_dict(results)
```

```python
age_input=float(input("Age : "))
bmi_input=float(input("BMI : "))
child_input=int(input("Children: "))
sex_input=int(input("Sex Male 0 or 1 : "))
smoker_s_input=int(input("Smoker yes 0 or 1 : "))
```

```python
future_prediction=grid.predict([[age_input,bmi_input,child_input,sex_input,smoker_s_input]])
print("Future Prediction is:{} ".format(future_prediction))
```
