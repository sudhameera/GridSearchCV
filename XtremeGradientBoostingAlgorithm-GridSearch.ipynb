```python
import pandas as pd
```

```python
data=pd.read_csv("insurance_pre.csv")
```

```python
data
```

```python
data=pd.get_dummies(data,drop_first=True)
```

```python
data
```

```python
independent=data[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
dependent=data[['charges']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBRegressor
grid_param={"subsample":[0.75, 1],  "colsample_bytree":[0.75, 1], "max_depth":[2, 6]}

grid=GridSearchCV(XGBRegressor(),grid_param, refit=True, verbose=3,n_jobs=-1)
grid.fit(x_train,y_train)
```

```python
results=grid.cv_results_
```

```python
results
```

```python
table=pd.DataFrame.from_dict(results)
```

```python
table
```

```python
print("The Rscore value with Best Tuning Parameter{}:".format(grid.best_params_))
```

```python
ypred=grid.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
rscore=r2_score(y_test,ypred)
```

```python
rscore
```
